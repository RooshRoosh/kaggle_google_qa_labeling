```python
len(urls['question_type_spelling'])
```

```python
res = []
for category in urls:
    for i, url in enumerate(urls[category]):
        r  = requests.get(url)
        soup = BeautifulSoup(r.text)

        title = soup.find('title').text
        question_body = soup.find('div', {'class': 'post-text'}).text
        answer = soup.find('div', {'class': 'answer accepted-answer'}).find('div', {'class': 'post-text'}).text
        
        res.append([title, question_body, answer, category])
        print(f'{category}: {i+1}/{len(urls[category])}')
        
res = pd.DataFrame(res, columns=['title', 'question_body', 'answer', 'category'])
```
